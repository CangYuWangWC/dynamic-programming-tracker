```python
import matplotlib.pyplot as plt
import numpy as np
import skimage.io as skio


image = skio.imread('examples/fruta.jpg')

image=image/255
plt.imshow(image)
plt.show()


```

```python
imagen_negra = np.zeros((120,140),dtype='uint8')

plt.imshow(imagen_negra,cmap=plt.get_cmap('gray'))

imagen_blanca = np.ones((120,140))
imagen_blanca[1,1]=0
plt.figure()
plt.imshow(imagen_blanca,cmap=plt.get_cmap('gray'))

for i in range(120):
    imagen_blanca[i,5]=0

    
plt.figure()
plt.imshow(imagen_blanca,cmap=plt.get_cmap('gray'))
plt.show()
```
